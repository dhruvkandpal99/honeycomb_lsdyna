```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
ex = pd.read_csv('experiment.csv')
true = pd.read_csv('true.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(ex[ex.columns[0]], ex[ex.columns[1]], label='Experimental', linewidth=2)
plt.plot(true[true.columns[0]], true[true.columns[1]], label='True', linestyle='--', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Experimental vs. True Data')
plt.legend()
plt.grid(True)
plt.show()
```
